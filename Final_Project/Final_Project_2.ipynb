```python
dummy_seed = pd.get_dummies(pre2['seed_number'], prefix = 'seed')
print dummy_seed.head()
```

```python
data = pre2.join(dummy_seed.ix[:,'seed_1':])
print data.head()
```

```python
data['intercept'] = 1.0
```

```python
train_cols = pre2.columns['seed_1']
logit = sm.Logit(data['ncaa_champ'], data[train_cols])
result = logit.fit()
```

```python
print result.summary()
```
